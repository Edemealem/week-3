```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

**Data Load**

```python
data = pd.read_csv(
    '/content/drive/MyDrive/10Academy/Week3/MachineLearningRating_v3.txt',
    delimiter='|')
data.head()
```

```python
data.columns
```

Handle missing value

```python
numeric_val = data.select_dtypes(include=['int64', 'float64']).columns
data[numeric_val] = data[numeric_val].fillna(data[numeric_val].mean())

categorical_val = data.select_dtypes(include=['object']).columns
data[categorical_val] = data[categorical_val].fillna(data[categorical_val].mode().iloc[0])
```

**Code for Data Segmentation**

```python
group_a = data[data['NewVehicle'] == 0]
group_b = data[data['NewVehicle'] == 1]
```

**Statistical Testing**

```python
# 1. Chi-squared test for risk differences across provinces
contingency_table_provinces = pd.crosstab(data['Province'], data['StatutoryRiskType'])
chi2_provinces, p_value_provinces, _, _ = stats.chi2_contingency(contingency_table_provinces)

# 2. Chi-squared test for risk differences between zip codes
contingency_table_zip = pd.crosstab(data['PostalCode'], data['StatutoryRiskType'])
chi2_zip, p_value_zip, _, _ = stats.chi2_contingency(contingency_table_zip)

# 3. T-test for profit margin differences between groups
# Calculate profit margins for both groups
group_a_profit = group_a['TotalPremium'] - group_a['TotalClaims']
group_b_profit = group_b['TotalPremium'] - group_b['TotalClaims']
t_stat_profit, p_value_profit = stats.ttest_ind(group_a_profit, group_b_profit)

# 4. Chi-squared test for gender risk differences
contingency_table_gender = pd.crosstab(data['Gender'], data['StatutoryRiskType'])
chi2_gender, p_value_gender, _, _ = stats.chi2_contingency(contingency_table_gender)
```

-  Analyze the p-value

```python
# Analyze p-values and print results
# Risk differences across provinces
if p_value_provinces < 0.05:
    print("Reject the null hypothesis: There are risk differences across provinces.")
else:
    print("Fail to reject the null hypothesis: No significant risk differences across provinces.")

# Risk differences between zip codes
if p_value_zip < 0.05:
    print("Reject the null hypothesis: There are risk differences between zip codes.")
else:
    print("Fail to reject the null hypothesis: No significant risk differences between zip codes.")

# Profit margin differences between groups
if p_value_profit < 0.05:
    print("Reject the null hypothesis: There are significant profit differences between groups.")
else:
    print("Fail to reject the null hypothesis: No significant profit differences between groups.")

# Risk differences between genders
if p_value_gender < 0.05:
    print("Reject the null hypothesis: There are risk differences between Women and Men.")
else:
    print("Fail to reject the null hypothesis: No significant risk differences between Women and Men.")
```
